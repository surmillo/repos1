# Установка jupyter notebook на Mac.  

 **1.** Нужно установить пакетный менеджер ***homebrew*** (утилита, c помощью которой далее можно устанавливать приложения). Для этого необходимо ввести строчку приведенную ниже команду в терминал на компьютере. Homebrew позволяет удобно управлять пакетами внутри операционной системы. С помощью одной строчкой устанавливать/удалять нужные приложения, запускать их непосредственно из терминала. 

```python
/bin/bash -c "$(curl -fsSL https://raw.githubusercontent.com/Homebrew/install/HEAD/install.sh)"
```

**2.** Устанавливаем пакет **python3** (стандартные библиотеки, необходимая документация и т. д. для нужной версии питона). Команду нужно также вводить в строку терминала (используя уже установленный homebrew). 

```python
brew install python3
```

**3.** С помощью команды pip устанавливаем непосредственно jupyter notebook. Jupyter — это программа-блокнот для записи и запуска кода. Ей можно пользоваться как средой разработки, причем от остальных сред разработки ее отличает то, что можно запускать отдельные части кода в 'произвольном' порядке, а также видеть выводы этих отдельных частей 

```python
pip install jupyter
```
